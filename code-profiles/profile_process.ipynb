## 1.1 Smoothing functions

Functions to smooth profiles and identify extrema.  The first minimum downstream of the thin rim bounds the fit domain.

```python
def fsmth_ind_first_min(y):
    """Smooths data and returns index of first minimum to left of global max, in smoothed data
    If no such minimum is found, return 0 (i.e., don't cut the fit domain)"""
    
    # List of indices of relative minima in smoothed-y
    y_smooth = fsmth_std_smooth(y)
    inds_relmin = sp.signal.argrelmin(y_smooth)[0]  # Unpack the tuple...
    
    if len(inds_relmin) is 0:
        return 0
    
    # Index of global (non-edge) maximum in y
    ind_max = fsmth_ind_max(y)
    
    # Find the largest value in inds_relmin, that's smaller than ind_max
    max_left_min = 0
    for ind_min in inds_relmin:
        if ind_min > max_left_min and ind_min < ind_max:
            max_left_min = ind_min
    
    return max_left_min


def fsmth_ind_max(y):
    """Smooths data and returns index of largest relative maximum in smoothed data"""
    ysmooth = fsmth_std_smooth(y)
    # Get indices of relative maxima
    inds_relmax = sp.signal.argrelmax(ysmooth)[0]  # Unpack the tuple...
    # Get index of the largest relative maximum
    best_ind = inds_relmax[0]
    for ind in inds_relmax[1:]:
        if ysmooth[best_ind] < ysmooth[ind]:
            best_ind = ind
    return best_ind


def fsmth_std_smooth(x):
    """Standardized smoothing -- SMOOTHING WINDOW CHOICE AFFECTS FIT DOMAIN!!!!!"""
    wlen = 21  # Arbitrary pick
    xsmooth = fsmth_smooth(x, window_len=wlen, window='hanning')
    padlen = (len(xsmooth) - len(x))/2  # Should be an integer
    return xsmooth[padlen:-padlen]


def fsmth_smooth(x, window_len=11, window='hanning'):
    """stolen from scipy cookbook (http://wiki.scipy.org/Cookbook/SignalSmooth)"""
    if window_len < 3:
        return x
    if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
        raise ValueError, "Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'"
    
    s = np.r_[x[window_len-1:0:-1],x,x[-1:-window_len:-1]]
    if window == 'flat': #  moving average
        w = numpy.ones(window_len,'d')
    else:
        w = eval('np.{}(window_len)'.format(window))
    
    y = np.convolve(w/w.sum(), s, mode='valid')
    return y
```

## 1.2 Fitting functions (built on `scipy.optimize.curve_fit`)

XSPEC-like functionality to freeze fit parameters

```python
def freeze_func(f, pars, ls_freeze):
    """Make frozen function
    Ex: if ls_freeze = [1,2,4], this builds expression:
    lambda x,p0,p3,p5,p6,p7 : f(x,p0,pars[1],pars[2],p3,pars[4],p5,p6,p7)
    """
    lambd_args = 'x,'
    lambd_body = 'f(x,'
    for i in xrange(len(pars)):
        if i in ls_freeze:
            lambd_body += 'pars[{:d}],'.format(i)
        else:
            lambd_args += 'p{:d},'.format(i)
            lambd_body += 'p{:d},'.format(i)
    
    lambd_args = lambd_args[:-1]
    lambd_body = lambd_body[:-1] + ')'
    
    # To explain the `f=f, pars=pars`, see http://stackoverflow.com/q/7349785
    # More than one way to skin a cat, though
    lambd_str = 'lambda ' + lambd_args + ', f=f,pars=pars : ' + lambd_body
    return eval(lambd_str)


def run_frz_fit(x, y, y_err, f, pars, ls_freeze, **kwargs):
    """Perform fit with frozen parameters
    func has form func(x, *pars)
    pars is a list of initial guesses
    ls_freeze is a list of parameter indices that are to be frozen.
    """
    
    if len(ls_freeze) > 0:
        fit_func = freeze_func(f, pars, ls_freeze)
        p_fit = [pars[i] for i in xrange(len(pars)) if i not in ls_freeze]
    else:
        fit_func = f
        p_fit = pars
    
    popt, pcov = run_fit(x, y, y_err, fit_func, p_fit, **kwargs)
    
    popt_aug = popt
    if len(ls_freeze) > 0:
        for i in ls_freeze:
            popt_aug = np.insert(popt_aug, i, pars[i])
    
    return popt, pcov, popt_aug


def run_fit(x, y, y_err, f, pars, **kwargs):
    """Convenience wrapper for sp.optimize.curve_fit"""
    return sp.optimize.curve_fit(f, x, y, p0=pars, sigma=y_err, absolute_sigma=True, **kwargs)


def print_fit_info(x, y, y_err, f, popt, pcov):
    """Pretty-prints fitting output"""
    np.set_printoptions(precision=2)
    print 'Chi-squared: {} with {} degrees of freedom'.format(chi2(x, y, y_err, f, popt), len(y)-len(popt))
    print 'Reduced chi-squared: {}'.format(chi2red(x, y, y_err, f, popt))
    print 'Errors: {}'.format(np.sqrt(np.diag(pcov)))
    #print 'Covariance matrix:', '\n', pcov
    np.set_printoptions(precision=8)  # Restore to default
```

### 1.3.1 Two exponential function with shared $x_0$, split at $x_s$

```python
def fit_2exp(x, y, y_err):
    """Fit data to a two-exponential function
    with initial guesses"""
    
    # Call signature: ff.two_exp(x, xs, x0, Cu, Cd, wu, wd, Au)
    f = ff.two_exp
    
    pars = ['xs', 'x0', 'Cu', 'Cd', 'wu', 'wd', 'Au']
    prnt_pars = [par + ' = {:5.4g}, ' for par in pars]
    prnt_pars = ''.join(prnt_pars)[:-2]  # Cut trailing comma, space
    
    #################
    # Initial guesses
    #################
    
    # Freeze: xs, x0
    xs = x[np.argmax(y)]  # Smoothing shifts peak slightly left
    x0 = xs + 0.5  # xs left of x0 in good fit (empirically speaking)
    # Free parameters: Cu, Cd, wu, wd, Au
    Cu = np.mean(y[-3:])  # The obvious guess
    Cd = np.mean(y[:3]) / 2.  # Usually shoots under trough
    wu = 0.5  # Upstream width
    wd = 2.5  # Downstream width
    Au = np.amax(y)/4.  # Upstream amplitude
    
    init_guess = [xs, x0, Cu, Cd, wu, wd, Au]
    
    print ('Initial guesses: ' + prnt_pars).format(*init_guess)
    
    ###################
    # First fit attempt
    ###################
    
    # Freeze xs, x0
    frz = [0, 1]
    popt, pcov, popt_all = run_frz_fit(x, y, y_err, f, init_guess, frz)
    print ('Fit (xs, x0 frozen): ' + prnt_pars).format(*popt_all)
    
    ####################
    # Second fit attempt
    ####################
    
    # Unfreeze: xs, x0; all parameters free
    init_guess = popt_all
    popt, pcov = run_fit(x, y, y_err, f, init_guess)
    print ('Fit (xs,x0,Cu,Cd thawed): ' + prnt_pars).format(*popt)
    
    #######################
    # Print fit information
    #######################
    
    print_fit_info(x, y, y_err, f, popt, pcov)
    
    return popt, pcov, f
```

### 1.3.4 Two exponential function, simplified ($x_s = x_0$), with manual cap imposed on function height...

Recall that the function is:

\begin{align*}
    h_{\mathrm{up}}(x) &= A_u \exp \left(\frac{x_0-x}{w_u}\right) + C_u \\
    h_{\mathrm{down}}(x) &= A_d \exp \left(\frac{x - x_0}{w_d}\right) + C_d
\end{align*}

Therefore, we demand that $h(x_0) = A_u + C_u = A_d + C_d$ be less than some cap.  If the sum $A_u + C_u$ is greater than this cap, then the function explodes everywhere.  So this sets a wall, essentially (though I think the function will tend to fit at the wall, if it wants to creep to steeper exponentials.

```python
def fit_2exp_simp_cap(x, y, y_err, hcap):
    """FIT THE PROFILE
    hcap indicates the fraction of the maximum data value
    at which to limit the function
    """
    
    # Very ad hoc way to impose height limit
    def func_capped(x, x0,Cu,Cd,wu,wd,Au):
        def f_help(x, x0,Cu,Cd,wu,wd,Au):
            if Au+Cu > hcap * np.amax(y):
                return 1e20
            return ff.two_exp_simp(x, x0,Cu,Cd,wu,wd,Au)
        f_vec = np.vectorize(f_help)
        return f_vec(x, x0,Cu,Cd,wu,wd,Au)
    
    # Call signature: two_exp_function_simp(x, x0, Cu, Cd, wu, wd, Au)
    f = func_capped
    
    pars = ['x0', 'Cu', 'Cd', 'wu', 'wd', 'Au']
    prnt_pars = [par + ' = {:5.4g}, ' for par in pars]
    prnt_pars = ''.join(prnt_pars)[:-2]  # Cut trailing comma, space

    #################
    # Initial guesses
    #################
    
    # Freeze: x0
    x0 = x[np.argmax(y)]
    # Free parameters: Cu, Cd, wu, wd, Au
    Cu = np.mean(y[-3:])  # The obvious guess
    Cd = np.mean(y[:3]) / 3.  # Usually shoots under trough
    wu = 0.5  # Upstream width
    wd = 2.5  # Downstream width
    Au = np.amax(y)/3.  # Upstream amplitude
    
    init_guess = [x0, Cu, Cd, wu, wd, Au]
    print ('Initial guesses: ' + prnt_pars).format(*init_guess)
    
    ###################
    # First fit attempt
    ###################
    
    frz = [0]
    popt, pcov, popt_all = run_frz_fit(x, y, y_err, f, init_guess, frz)
    print ('Fit (x0 frozen): ' + prnt_pars).format(*popt_all)
    
    ####################
    # Second fit attempt
    ####################
    
    # Unfreeze: xs, x0; all parameters free
    init_guess = popt_all
    popt, pcov = run_fit(x, y, y_err, f, init_guess)
    print ('Fit (x0 thawed): ' + prnt_pars).format(*popt)
    
    #######################
    # Print fit information
    #######################
    
    print_fit_info(x, y, y_err, f, popt, pcov)
    
    return popt, pcov, f
```

## 2.1 FWHM utility functions, stretching functions

```python
def find_fmax(a, b, f, pars):
    """LOCAL maximum of a scalar function, within a given interval"""
    # Multiply by -1 to maximize
    res = sp.optimize.minimize_scalar(lambda x: -1*f(x, *pars), bounds=(a, b), method='Bounded')
    return res.x

def get_fwhm(a, b, f, pars, x_max=None, y_cap=None):
    """Manually calculate FWHM given a bracketing interval, location of maximum, and function.
    Assumptions -- there exists a single local maximum, from which the function
    MONOTONICALLY DECREASES on both sides
    
    If y_max is supplied, use THAT instead of actual maximum!
    Then it's no longer a true FWHM for the function, but may be better
    estimate of FWHM for the data
    
    Inputs:
        a, b (float): bounding interval
        f (function): scalar function in first argument, with form f(x, *popt)
        pars (list): list of parameters for function f
        x_max (float): location of maximum if known (skip maximization step)
        y_cap (float): fixed "maximum" value, to get full-width-at-half-"y_cap" instead
    
    Output:
        3-tuple of FWHM and left, right locations of FWHM measurement
    """
    if x_max is None:
        x_max = find_fmax(a, b, f, pars)
    
    if y_cap is None:
        y_hm = f(x_max, *pars) / 2.
    else:
        y_hm = y_cap / 2.
    
    y_lims = f(np.array([a,b]), *pars)  # Assumes that f is vectorized in x
    if any(y_lims >= y_hm):  # Function must fall below half max on both sides
        print 'WARNING: cannot calculate FWHM, peak not resolved in interval'
        return float('NaN'), float('NaN'), float('NaN')

    fwhm_left = sp.optimize.brentq(lambda x: f(x, *pars) - y_hm, a, x_max)
    fwhm_right = sp.optimize.brentq(lambda x: f(x, *pars) - y_hm, x_max, b)
    return fwhm_right - fwhm_left, fwhm_left, fwhm_right


def chi2(x, y, y_err, f, pars):
    """Chi-squared for a scalar function fitted to data with y-errors
    
    Inputs:
        x, y, y_err (iterable): the usual suspects
        f (function): function with call signature f(x, *pars)
        pars (list):  list of parameters for function f
    Output:
        chi-squared value (float)
    """
    return np.sum( ((y - f(x, *pars))/y_err)**2 )


def chi2red(x, y, y_err, f, pars):
    """Reduced chi-squared for a scalar function fitted to data with y-errors
    
    N.B. for a non-linear model fit, assigning 1 degree of freedom
    to each parameter may be an OVERestimate?
    (there may be other caveats/issues at hand)
    
    Inputs:
        x, y, y_err (iterable): the usual suspects
        f (function): function with call signature f(x, *pars)
        pars (list):  list of parameters for function f
    Output:
        reduced chi-squared value (float)
    """
    return chi2(x, y, y_err, f, pars) / (len(y) - len(pars))
```

Stretching functionality

```python
def stretch_x(x, xi, x0, xconst=50):
    """Parabolic stretch centered on x=x0, fixed points at x=0, x=x0
    Follows Ressler et al. [2014]
    
    Constant here is 50 arcsec instead of 200 arcsec.
    Doesn't matter, mainly scales the input range of xi
    
    No error checking -- you put bad xi, you get bad stretch
    
    Inputs:
        x (float / iterable): original x-values
        xi (float): stretch factor
        x0 (float): fixed point of stretch
    Output:
        stretched version of x (float/iterable depending on input)
    """
    return x * (1 + xi * (x - x0)/(xconst - x0))

def safe_stretch(xi, x0, xconst=50):
    """Check if xi is in the safe range
    I.e., the range in which
    1. stretch works as desired on interval [0, xconst]
    2. stretch is 1-to-1 on said interval
    """
    if x0 > xconst:
        raise Exception('Invalid stretch x0={} > xconst'.format(x0))
    return abs(xi) < (xconst - x0)/x0

def inv_stretch_x(xs, xi, x0, xconst=50):
    """Calculates the desired value of x, given a stretched x
    Returns the "sensible" value for our stretch
    
    If a=xi >0, then positive root is bigger, positive (2nd root)
    If a=xi <0, then positive root is smaller (1st root)
    
    Can return a NaN value if stretch cannot be inverted
    (physical meaning: stretch breaks down at large distances)
    Vectorized in xs, by nature (but not in other arguments)
    """
    a = xi
    b = xconst - x0*(1 + xi)
    c = -(xconst - x0) * xs
    d = b**2 - 4*a*c
    posroot = (-b + np.sqrt(d)) / (2*a)
    
    if not safe_stretch(xi, x0, xconst):
        if xi < 0:
            raise Exception('xi={} < {}'.format(xi))
        else:
            print 'Warning: possibly bad inverse stretch (xi too large)'
    
    if xi == 0:
        return xs
    return posroot

def stretch_func(f, xi, x_max):
    """Stretched version of a function with call signature f(x, *pars)
    If function f is vectorized in x, output function will also be vectorized"""
    return lambda x, *pars: f(stretch_x(x, xi, x_max), *pars)
```

## 2.3 Functions to estimate uncertainty

```python
def get_stretch_limits(x, y, y_err, f, pars, xmax):
    """Limits on stretch parameter xi for 90% confidence interval
    x-coordinate stretch function follows Ressler et al.
    pars specifies least-squares best fit parameters
    """
    
    def chi2_threshold(xi):
        """Threshold for chi^2 uncertainty. Output > 0 if \Delta\chi^2 > 2.7"""
        def chi2_stretch(xi):
            return chi2(x, y, y_err, stretch_func(f, xi, xmax), pars)
        chi2_stretch = np.vectorize(chi2_stretch)
        
        chi2_min = chi2(x, y, y_err, f, pars)  # same as chi2_stretch(xi=0)
        
        return chi2_stretch(xi) - chi2_min - 2.7  # Magic number...
    
    # Increasing positive xi (decreasing FWHM) uncertainty
    # Empirically, positive xi is better behaved than negative xi
    xi_max = 0.01
    while chi2_threshold(xi_max) < 0:
        xi_max += 0.01
    xi_bound_pos = sp.optimize.brentq(chi2_threshold, 0, xi_max)
    if not safe_stretch(xi_bound_pos, xmax):
        print 'WARNING: xi may be too large, pathological behavior may occur for small x'

    # Decreasing negative xi (increasing FWHM) uncertainty
    # Empirically, this can blow up quickly (stretch becomes meaningless if too far)
    xi_min = -0.01
    while chi2_threshold(xi_min) < 0:
        xi_min -= 0.01
    xi_bound_neg = sp.optimize.brentq(chi2_threshold, xi_min, 0)
    if not safe_stretch(xi_bound_neg, xmax):
        raise Exception('ERROR: xi negative and too large, pathological behavior WILL OCCUR (flipping)')
    
    return xi_bound_neg, xi_bound_pos


def get_fwhm_errs(fwhm, fwhm_left, fwhm_right, xi_neg, xi_pos, x0):
    """FWHM limits based on stretch limits (in x-coordinate only)
    Applies an inverse stretch operation to transform
    old coordinates to new coordinates
    
    E.g., f_stretch(new_fwhm_b) = f(x'(new_fwhm_b)) = f(old_fwhm_b)
    So we solve the inverse problem old_fwhm_b = x'(new_fwhm_b)
    
    Inputs:
        fwhm: fwhm
        fwhm_a, fwhm_b: x coordinate bounds on fwhm location
        xi_neg, xi_pos: some confidence limits on xi
        x0: center for xi-stretch
    Output:
        errors on FWHM
    """
    intrv = np.array([fwhm_left, fwhm_right])
    
    min_intrv = inv_stretch_x(intrv, xi_pos, x0)
    fwhm_min = min_intrv[1] - min_intrv[0]
    
    max_intrv = inv_stretch_x(intrv, xi_neg, x0)
    fwhm_max = max_intrv[1] - max_intrv[0]
    
    err_neg = fwhm - fwhm_min
    err_pos = fwhm_max - fwhm
    
    return err_neg, err_pos
```

## 2.4 Get everything at once (because I'm lazy)

```python
def get_fwhm_all(x, y, y_err, f, pars, y_cap=None):
    """After running a fit routine, compute useful numbers associated with FWHM"""
    x0 = x[np.argmax(y)]
    xmax = find_fmax(x0-5, x0+5, f, pars)  # Search within +/- 5 arcsec. of x0 -- empirical MAGIC NUMBER
    # Note that this is the FUNCTION's xmax, not the DATA xmax
    fwhm, fwhm_a, fwhm_b = get_fwhm(np.amin(x), np.amax(x), f, pars, x_max=xmax, y_cap=y_cap)
    xi_neg, xi_pos = get_stretch_limits(x, y, y_err, f, pars, xmax)
    fwhm_errs = get_fwhm_errs(fwhm, fwhm_a, fwhm_b, xi_neg, xi_pos, xmax)
    
    return fwhm, (fwhm_a, fwhm_b), fwhm_errs
```

## 3.1 Set-up: build initial data structure for regions and energy bands

Data values, uncertainties, etc. are imported/calculated here.  E.g., to use total counts, average counts, intensity flux, change code here.

```python
def init_region_dict(ds9physreg, inroot, ctroot, labels):
    """Initialize region dictionary structure, to hold information about
    regions and energy bands.
    
    Start by adding data with errorbars, region information (dimensions).
    Inputs: filenames (strings)
    Output: dictionary, keyed by region number (values are more dictionaries)
    """
    
    def get_proj_dims(regstrs):
        """Get projection dimensions from list of regions, must all be projections
        Input: list of region strings
        Output: list of 2-tuples
            1st element is length (radial distance)
            2nd element is thickness (dist. along which projection integrates)

        WARNING: DIMENSIONS ARE NOT THE SAME.  Length is in ARCSEC, thickness is in PIXELS / PHYSICAL UNITS.
        MAGIC NUMBER SET HERE -- CHANDRA SPATIAL RESOLUTION
        """
        pixel2arcsec = 0.5  # Chandra spatial resolution (approx -- we are 4 arcmin. off axis if that matters?)

        dims = []
        for line in regstrs:
            rtype, rnums, rprops = regparse.regparse(line)
            if 'projection' not in rtype:
                raise Exception('Projection region expected; region type was {}'.format(rtype))
            x1,y1,x2,y2,w = rnums
            length = np.sqrt((x2-x1)**2 + (y2-y1)**2) * pixel2arcsec
            dims.append((length, w))
        return dims
    
    
    # Get radial lengths of each region (convert physical pixels to arcseconds)
    dims = get_proj_dims(regparse.load_ds9reg(ds9physreg))

    regions = {}
    
    for i in xrange(len(dims)):
        # Set up each region's dictionary
        reg = regions[i+1] = {}  # Modifying reg, modifies regions[i+1]
        
        reg['length'] = dims[i][0]  # Units: arcsec
        reg['thickness'] = np.floor(dims[i][1])  # Units: PIXELS (apply floor to be consistent with DS9)
        print ('Region {:02d} length (arcsec): {: 0.2f}, '
               'thickness (pixels): {: 0.2f}').format(i+1, reg['length'], reg['thickness'])

        for lab in labels:

            # Set up each energy band's dictionary
            reg[lab] = {}

            # Import intensity data
            infile = '{0}_{1:02d}_band_{2}.dat'.format(inroot, i+1, lab)
            data = np.loadtxt(infile)
            x = (data[:,0] - 1) * reg['length'] / (len(data[:,0]) - 1)  # Convert to arcsec
            y = data[:,1]

            ctfile = '{0}_{1:02d}_band_{2}.dat'.format(ctroot, i+1, lab)

            # Import count data
            data_cts = np.loadtxt(ctfile)
            cts = data_cts[:,1]  # Counts averaged over region thickness
            cts_err = np.sqrt(cts * reg['thickness']) / reg['thickness']  # More explicit calculation
            cts_err[cts_err==0] = np.sqrt(1) / reg['thickness']  # Give 0 counts appropriate non-zero error
            
            # Use count data to compute y errors
            y_err = cts_err / cts * y
            # Edge cases -- if cts=0, or y=0 (since cts_err is always nonzero)
            y_err[cts==0] = y[cts==0]  # If cts==0, use y value: same as 1 +/- 1 for Poisson error            
            y_err[y_err==0] = np.amin(y_err[np.nonzero(y_err)])  # If y=0 (even if cts!=0), use min non-zero error

            # Save to energy band dictionary
            reg[lab]['data'] = x, y, y_err

            # Apply fit domain cuts
            cut = fsmth_ind_first_min(y)
            reg[lab]['data-fit'] = x[cut:], y[cut:], y_err[cut:]
    
    return regions
```

```python
def fill_region_dict(reg):
    """reg should be an input region dictionary, to be filled with data -- FWHMs and other things
    Spews out a crapton of information and plots
    Relies on several variables in the global namespace -- be careful.
    """
    
    fig1, axes1 = plt.subplots(1, len(labels), figsize=(6*len(labels),5))
    fig2, axes2 = plt.subplots(1, len(labels), figsize=(6*len(labels),5))
    
    for lab, ax1, ax2 in zip(labels, axes1, axes2):
        print '--------------'
        print 'Band: {}'.format(lab)
        print '--------------'
        
        x_fit, y_fit, y_err_fit = reg[lab]['data-fit']
        
        ###################
        # Main fit and FWHM
        ###################
        
        popt, pcov, f = fit_2exp_simp(x_fit, y_fit, y_err_fit)
        reg[lab]['f'] = f
        reg[lab]['pars'] = popt
        
        # For background subtraction, replace `f` with
        # lambda x, *pars: f(x,*pars) - y_bkg
        w, lims, errs = get_fwhm_all(x_fit, y_fit, y_err_fit, f, popt, y_cap=None)
        reg[lab]['fwhm'] = w
        reg[lab]['fwhm-lims'] = lims
        reg[lab]['fwhm-errs'] = errs
        
        ###########################
        # Additional fits and FWHMs
        ###########################
        
        # Capped FWHM for fit function; maximum set to largest data point
        w, lims, errs = get_fwhm_all(x_fit, y_fit, y_err_fit, f, popt, y_cap=np.amax(y_fit))
        reg[lab]['fwhmc'] = w
        reg[lab]['fwhmc-lims'] = lims
        reg[lab]['fwhmc-errs'] = errs
        
        # FWHM for a spline fit
        # Cannot use get_fwhm_all, because errors will be horrendous/bad
        s = fit_spline(x_fit, y_fit, y_err_fit, sfactor=0.05)
        w, lims, errs = get_fwhm_all(x_fit, y_fit, y_err_fit, s, [], y_cap=None)  # Errors are meaningless
        reg[lab]['fwhms'] = w
        reg[lab]['fwhms-lims'] = lims
        
        # Fit with a CAP on amplitude (fcap)
        popt_fitc, pcov_fitc, f_fitc = fit_2exp_simp_cap(x_fit, y_fit, y_err_fit, hcap=1.0)
        # Then get FWHM as usual (y_cap=None)
        w, lims, errs = get_fwhm_all(x_fit, y_fit, y_err_fit, f_fitc, popt_fitc, y_cap=None)
        reg[lab]['fwhm_fitc'] = w
        reg[lab]['fwhm_fitc-lims'] = lims
        reg[lab]['fwhm_fitc-errs'] = errs
        
        ############################
        # Plot the data, fits, FWHM,
        # and FWHM left/right bounds
        ############################
        x, y, y_err = reg[lab]['data']
        ax1.errorbar(x, y, yerr=y_err, fmt='rx', alpha=0.2)  # All raw data
        ax1.plot(x, fsmth_std_smooth(y), '-k', alpha = 0.2)  # Show smoothing to set fit domain
        ax1.errorbar(x_fit, y_fit, yerr=y_err_fit, fmt='bo', alpha=0.7)  # Show data used in fit

        x_m = np.linspace(x_fit[0], x_fit[-1], num=200)  # x values for best fit model
        ax1.plot(x_m, f(x_m, *popt), '-k', alpha=1)  # Best fit model

        ax1.set_xlabel('Radial dist. (arcsec.)')
        ax1.set_ylabel('Est. counts ~ intensity')
        ax1.set_title(lab)
        
        ############################
        # Plot FWHM bounds near peak
        ############################
        
        ax2.errorbar(x_fit, y_fit, yerr=y_err_fit, fmt='bo', alpha=0.2)  # Data used in fit
        ax2.plot(x_m, f(x_m, *popt), '-k', alpha=1)  # Best fit model
        ax2.plot(x_m, s(x_m), '-r', alpha=0.6)  # Interpolating spline
        
        ax2.set_xlim(x_fit[np.argmax(y_fit)]-0.5, x_fit[np.argmax(y_fit)]+0.5)

        # Delineate FWHM measurements, if possible
        if np.isfinite(reg[lab]['fwhm']):
            a, b = reg[lab]['fwhm-lims']
            ax2.axvline(a, alpha=1, ls='-', c='k', lw=2)
            ax2.axvline(b, alpha=1, ls='-', c='k', lw=2)
            ax2.set_xlim(min(ax2.get_xlim()[0], a-2),
                         max(ax2.get_xlim()[1], b+2))
        if np.isfinite(reg[lab]['fwhmc']):
            a, b = reg[lab]['fwhmc-lims']
            ax2.axvline(a, alpha=1, ls='--', c='g', lw=2)
            ax2.axvline(b, alpha=1, ls='--', c='g', lw=2)
            ax2.set_xlim(min(ax2.get_xlim()[0], a-2),
                         max(ax2.get_xlim()[1], b+2))
        if np.isfinite(reg[lab]['fwhms']):
            a, b = reg[lab]['fwhms-lims']
            ax2.axvline(a, alpha=1, ls=':', c='r', lw=2)
            ax2.axvline(b, alpha=1, ls=':', c='r', lw=2)
            ax2.set_xlim(min(ax2.get_xlim()[0], a-2),
                         max(ax2.get_xlim()[1], b+2))
        
        if ax2.get_xlim()[0] == x_fit[np.argmax(y_fit)]-0.5:
            ax2.set_xlim(x_fit[0], x_fit[-1])
        ax2.set_xlabel('Radial dist. (arcsec.)')
        ax2.set_ylabel('Est. counts ~ intensity')
    
    #############################
    # Plot FWHM energy dependence
    #############################

    plt.figure(figsize=(8,6))
    
    for lab, m in zip(labels, xrange(len(labels))):
        fwhm, fwhmc, fwhms = reg[lab]['fwhm'], reg[lab]['fwhmc'], reg[lab]['fwhms']
        fwhm_fitc = reg[lab]['fwhm_fitc']
        
        # Plot FWHMs calculated from different methods, if possible
        if np.isfinite(fwhm):
            nerr, perr = reg[lab]['fwhm-errs']
            plt.errorbar([m+1], [fwhm], yerr=[[nerr], [perr]], fmt='bs')  # Straight 2-exp fit
        if np.isfinite(fwhmc):
            nerr, perr = reg[lab]['fwhmc-errs']
            plt.errorbar([m+1+0.03], [fwhmc], yerr=[[nerr], [perr]], fmt='go', mfc='none')  # 2-exp fit, cap on FWHM
        if np.isfinite(fwhms):
            plt.plot([m+1+0.06], fwhms, 'rv')  # Spline
        if np.isfinite(fwhm_fitc):
            nerr, perr = reg[lab]['fwhm_fitc-errs']
            plt.errorbar([m+1-0.03], [fwhm_fitc], yerr=[[nerr], [perr]], fmt='b^', mfc='none')  # 2-exp capped fit

    plt.legend(['2-exp', '2-exp capped fwhm', 'spline', '2-exp capped fit'], loc='best')  # Legend not always correct
    plt.xlim([0,len(labels)+1])
    plt.xlabel('Energy band number')
    plt.ylabel('FWHM (arcsec.)')
    plt.show()
    print ''
```
