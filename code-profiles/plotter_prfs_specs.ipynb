```python
for n in regions.keys():
    reg = regions[n]
    
    print ''
    print '---------'
    print 'Region {:02d}'.format(n)
    print '---------'
    
    ########################
    # FWHMs and profile fits
    # for all energy bands
    ########################
    
    fig, axes = plt.subplots(1, len(labels), figsize=(6*len(labels),4.5))
    
    for lab, ax in zip(labels, axes):
        # Add raw data
        x, y, e = reg[lab]['data']
        xf, yf, ef = reg[lab]['data-fit']
        cut = reg[lab]['cut']
        # Fit domain data
        ax.errorbar(xf, yf, ef, fmt='ko', capsize=0, alpha=0.8)
        # Outside fit domain data
        ax.errorbar(x[:cut], y[:cut], e[:cut], fmt='r.', capsize=0, alpha=0.5)
        
        # Add FWHMs, fit function if FWHM could be found and was not blacklisted
        if np.isfinite(reg[lab]['meas']['fwhm']):
            # Define fit function and add to namespace.  Warning: string parsing may break easily
            f_src = reg[lab]['f']
            f_def = f_src.split('\n')[0]  # Get first line of src
            f_name = re.split(r'[\s(]+', f_def)[1]  # Get function name
            exec(f_src)
            f = eval(f_name)
            
            # Plot fit function
            x_m = np.linspace(np.amin(xf), np.amax(xf), 100)
            ax.plot(x_m, f(x_m, *reg[lab]['pars']), '-b')
            
            # Plot FWHM limits
            fwhm_a, fwhm_b = reg[lab]['meas']['fwhm-lims']
            ax.axvline(fwhm_a, alpha=1, ls='--', c='k', lw=1)
            ax.axvline(fwhm_b, alpha=1, ls='--', c='k', lw=1)
            
            # Plot FWHM limits for CAPPED fwhm
            fwhm_a, fwhm_b = reg[lab]['meas']['fwhmc-lims']
            ax.axvline(fwhm_a, alpha=0.7, ls='-', c='g', lw=1)
            ax.axvline(fwhm_b, alpha=0.7, ls='-', c='g', lw=1)
        
        ax.axis('tight')
        ax.set_ylim(0., ax.get_ylim()[1])
        ax.set_xlabel('Radial position (arcsec.)')
        ax.set_ylabel('Intensity (A.U.)')
        ax.set_title(lab)
    
    plt.show()
    
    
    #####################################
    # Show spectra and extraction regions
    #####################################
    
    fig = plt.figure(figsize=(18,5))
    
    # ------------------
    # EXTRACTION REGIONS
    # ------------------
    
    ax = plt.subplot2grid((5,3), (0,0), rowspan=5)
    
    # Plot data from highest energy band
    lab = labels[-1]
    x, y, e = reg[lab]['data']
    xf, yf, ef = reg[lab]['data-fit']
    cut = reg[lab]['cut']
    # Fit domain data
    ax.errorbar(xf, yf, ef, fmt='ko', capsize=0, alpha=0.8)
    # Outside fit domain data
    ax.errorbar(x[:cut], y[:cut], e[:cut], fmt='r.', capsize=0, alpha=0.5)
    
    # Plot extraction locations
    for xcut in reg['info']['spec_cuts']:
        if xcut < 0.05:
            xcut = 0.05
        ax.axvline(xcut, alpha=1, ls='--', c='b', lw=2)
    
    # Set plot axes/labels
    ax.axis('tight')
    ax.set_ylim(0., ax.get_ylim()[1])
    ax.set_xlabel('Radial position (arcsec.)')
    ax.set_ylabel('Intensity (A.U.)')
    
    # ----------------------------
    # Upstream, downstream spectra
    # ----------------------------
    
    for sect, clr, col in zip(['down', 'up'], ['r', 'k'], [1,2]):
        
        data = reg['info']['specs'][sect]['data']
        for key in data.keys():
            exec("{k} = data['{k}']".format(k = key))
        res = y - m  # Residuals
        
        # Plot spectrum
        ax = plt.subplot2grid((5,3), (0,col), rowspan=4)
        ax.errorbar(x, y, xerr = xE, yerr = yE, fmt='.'+clr, capsize=0, alpha=0.5)
        ax.plot(x, m, '-'+clr, alpha=0.5)
        ax.set_xscale('log')
        ax.set_yscale('log')
        ax.axis('tight')
        ax.set_ylabel(r'Normalized counts s$^{-1}$ keV$^{-1}$')
        
        plt.setp(ax.get_xticklabels(), visible=False)  # Share x-axis with resid plot
        
        # Plot residuals
        ax2 = plt.subplot2grid((5,3), (4,col), sharex=ax)  # Share axis
        ax2.errorbar(x, res, xerr=xE, yerr=yE, fmt='.'+clr, capsize=0, alpha=0.5)
        ax2.axhline(0, c='k')
        
        ax2.set_xscale('log')
        ax2.axis('tight')
        ax2.set_ylim(-1*max(np.abs(ax2.get_ylim())),max(np.abs(ax2.get_ylim())))  # Symmetrize y-limits
        ax2.get_yaxis().get_major_formatter().set_powerlimits((-2, 4))  # Force scientific notation
        
        ax2.set_xlabel('Energy (keV)')
    
    # For the spectra/residuals to share axes
    fig.subplots_adjust(hspace=0)
    
    plt.show()
```
