```python
def init_tab_vs(fexpr, vsvals):
    """vsvals must be sorted and match filenames"""
    tabs = {}
    for fname, vs in zip(sorted(glob(fexpr)), vsvals):
        with open(fname, 'r') as fpkl:
            tabs[vs] = pickle.load(fpkl)
    return tabs

def init_tab_ab(fexpr, abvals):
    """Read in tables for list of ab values, given glob expr
    Assumes ab values uniquely specified in filename with {:0.2e}"""
    tabs = {}
    flist = glob(fexpr)
    for ab in abvals:
        abstr = '{:0.2e}'.format(ab)
        tabmatch = [x for x in flist if abstr in x]
        if len(tabmatch) > 1:  # Not expected
            print 'warning, found more than one match to given ab value'
            print 'taking first one: {}'.format(tabmatch[0])
        with open(tabmatch[0], 'r') as fpkl:
            tabs[ab] = pickle.load(fpkl)
    return tabs

def init_vs(f_veloc, d3kpc=None):
    """Read in interpolated shock velocities, convert to dict format
    If d3kpc is specified, rescale appropriately"""
    arr = np.loadtxt(f_veloc)
    
    if d3kpc is not None:
        arr = arr * d3kpc  # arr assumed generated at d3kpc
    
    return dict(zip(xrange(1, len(arr)+1), arr))

def init_eta2(f_eta2, mu_vals=None):
    arr = np.loadtxt(f_eta2)
    
    if mu_vals is None:
        mu_vals = [0, 1./3, 0.5, 1, 1.5, 2]
    assert len(mu_vals) == len(arr[0,1:])
    # Currently not using mu values for anything
    
    return {n: row[1:] for n, row in zip(xrange(1, len(arr)+1), arr)}
```

# Using table parameters, compute FWHMs at 0.01 keV

This allows us to threshold the strong damping cases.  Let's see what comes out.

```python
%%capture

kev_ext = np.array([0.01])  # Compute new FWHM at 0.01 keV
mu = 1.0

savefname = 'backup-tab-ext-141111-bmin_5e-6.pkl'

TYCHO_TABS_DAMP_EXT = {}

for ab in TYCHO_AB_VALS:
    
    TYCHO_TABS_DAMP_EXT[ab] = {}
    TYCHO_TABS_DAMP_EXT[ab][mu] = {}
    
    for eta2, (B0_vals, fwhms) in TYCHO_TABS_DAMP[ab][mu].items():
        
        fwhms_ext = []
        
        for B0 in B0_vals:
            new_fwhm = models.full_width(snrcat.make_tycho(), kev_ext, mu, eta2, B0,
                                         idamp=True, damp_ab=ab, damp_bmin=5e-6, verbose=False)
            fwhms_ext.append(new_fwhm[0])
        
        TYCHO_TABS_DAMP_EXT[ab][mu][eta2] = (B0_vals, np.array(fwhms_ext))
        
    with open(savefname, 'w') as fpkl:  # Save after every ab value
        pickle.dump(TYCHO_TABS_DAMP_EXT, fpkl)
```

Now visualize it

```python
for ab in TYCHO_AB_VALS:
    print 'ab = {:0.3f}'.format(ab)
    plt.figure(figsize=(12,8))
    
    bounds = []
    bounds_sharp = []
    
    for eta2, (B0_vals, fwhms) in TYCHO_TABS_DAMP_EXT[ab][1.0].items():
        
        msk = fwhms < 0.99 * snrcat.make_tycho().rsarc
        B0_vals_fwhm = B0_vals[msk]
        B0_vals_nofwhm = B0_vals[np.logical_not(msk)]
        
        sortmsk = np.argsort(B0_vals_fwhm)
        
        if any(B0_vals_fwhm):
            B0_min = B0_vals_fwhm[sortmsk][0]  # Get the smallest B0 value that yields valid FWHM at 0.01 keV
            fwhm_max = fwhms[msk][sortmsk][0]  # Get the FWHM corresponding to smallest valid B0 value
        else:
            B0_min = np.nan
            fwhm_max = np.nan
        
        bounds.append((eta2, 1e6 * B0_min, fwhm_max))
        
        if any(B0_vals_nofwhm):
            bounds_sharp.append(1e6*np.sort(B0_vals_nofwhm)[-1])
        else:
            bounds_sharp.append(np.nan)
        
        plt.scatter(eta2*np.ones(len(B0_vals_fwhm)), B0_vals_fwhm*1e6, c='g', s=10, lw=0, alpha=0.5)
        plt.scatter(eta2*np.ones(len(B0_vals_nofwhm)), B0_vals_nofwhm*1e6, c='r', s=10, lw=0, alpha=0.5)
    
    bounds = np.array(bounds).T
    np.set_printoptions(precision=2, suppress=True, threshold=10)
    srt = np.argsort(bounds[0,:])  # srt by eta2
    print 'eta2:\t\t', bounds[0,srt]
    print 'B0 min (good):\t', bounds[1,srt]
    print 'FWHM max:\t', bounds[2,srt]
    
    if any(bounds_sharp):
        bounds_sharp = np.array(bounds_sharp)
        print 'B0 min (bad):\t', bounds_sharp[srt]
    
    plt.xscale('log')
    fplot(r'$\eta_2$', r'$B_0$ (microG)', axargs='tight', scales=('log', 'linear'))
    plt.show()
```

```python
#for eta2 in [0.001, 0.01, 0.1, 1.0, 10, 100, 100]:
for eta2 in [1]:
    parspace_fname = '../../../Desktop/pargrid-eta2_{}-model'.format(eta2)
    
    params = [( eta2, float('inf'), [20e-6, 30e-6, 100e-6, 400e-6]),
              ( eta2, 0.2,  [20e-6, 30e-6, 100e-6, 400e-6]),
              ( eta2, 0.05,  [20e-6, 30e-6, 100e-6, 400e-6]),
              ( eta2, 0.02,  [20e-6, 30e-6, 100e-6, 400e-6]),
              ( eta2, 0.005, [20e-6, 30e-6, 100e-6,{} 400e-6]),
              ( eta2, 0.002, [20e-6, 30e-6, 100e-6, 400e-6])]

    kevs = np.array([4.136e-9, 1.0, 4.0])
    fmts = ['-r', '-g', '-b']

    kws = {'irmax':400, 'iradmax':200, 'rminarc':40, 'irad_adapt':False,
           'damp_bmin': 5e-6}
    snr = snrcat.make_tycho()

    nrows, ncols = len(params), max(map(lambda x: len(x[2]), params))
    fig, axes = plt.subplots(nrows, ncols, sharex=True, sharey=True,
                             figsize=(4*ncols, 3*nrows))

    for row_pars, ax_list in zip(params, axes):

        eta2, ab, B0_list = row_pars
        for B0, ax in zip(B0_list, ax_list):

            idamp_flag = ab < 1.0
            _ = models.full_width(snr, kevs, 1.0, eta2, B0,
                                  get_prfs=True, idamp=idamp_flag, damp_ab=ab,
                                  **kws)
            fwhms, igrids, rgrids = _

            # Plot all the data
            for i in xrange(len(kevs)):
                ind_pk = argrelmax(igrids[i])[0]
                if kevs[i] > 0.1:  # X-ray energies
                    pk_scale = np.amax(igrids[i]) if len(ind_pk) == 0 else igrids[i][ind_pk[-1]]
                    ax.plot(rgrids[i], igrids[i] / pk_scale, fmts[i], alpha=0.7, label=r'{:0.2g} keV'.format(kevs[i]))
                else:  # Radio
                    pk_scale = np.amax(igrids[i])
                    ax.plot(rgrids[i], igrids[i] / pk_scale, fmts[i], label=r'{:0.2g} GHz'.format(kevs[i] * 2.418e8))
                    # 1 keV/h = 2.418e8 GHz

            # Standard subplot settings and annotations
            ax.minorticks_on()
            if idamp_flag:
                ax.axvspan((1-ab)*snr.rsarc, snr.rsarc, color='0.2', alpha=0.2)  # Shade in damping region
            ax.set_xlim(ax.get_xlim()[0], snr.rsarc)
            ax.set_ylim(0., 1.01)
            ax.grid(color='k', which='major', alpha=0.5, linestyle=':')  # Gridding
            ax.axhline(y=0.5, alpha=0.5, color='k')  # Half-maximum line

            # Target certain subplots for legend, plot axes, etc
            if row_pars is params[0]:  # First row of grid
                col_lab = r'$B_0 = {:0.0f}'.format(B0*1e6) + r'\;\mu\mathrm{G}$'
                ax.text(0.5, 1.1, col_lab, fontsize=16, transform=ax.transAxes, ha='center')
                if ax is ax_list[0]:  # First row, first column
                    corner_lab = r'$\eta_2 = {:0.3g}$'.format(eta2)
                    ax.text(-0.5, 1.1, corner_lab, fontsize=20, transform=ax.transAxes, va='center',
                            color='red')

            if ax is ax_list[0]:  # First column of grid
                row_lab = r'$a_b = {:0.3f}$'.format(ab) if idamp_flag else r'$a_b = \infty$'+'\n(loss-limited)'
                ax.text(-0.5, 0.5, row_lab, fontsize=16, transform=ax.transAxes, va='center')
                ax.set_ylabel('Intensity (a.u.)')

            if ax in axes[-1, :]:  # Bottom row of subplots
                ax.set_xlabel('Radial position (arcsec.)')

            if B0 is B0_list[-1]:  # Last plot of each row
                ax.legend(loc='upper left', prop={'size':8})

    plt.tight_layout()
    plt.savefig(parspace_fname + '.pdf')
    plt.show()
```
