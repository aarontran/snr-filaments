Some code for table displays in iPython notebooks ([source](http://calebmadrigal.com/display-list-as-table-in-ipython-notebook/)).

```python
class ListTable(list):
    """ Overridden list class which takes a 2-dimensional list of 
    the form [[1,2,3],[4,5,6]], and renders an HTML Table in ipynb.
    Source: http://calebmadrigal.com/display-list-as-table-in-ipython-notebook/
    """
    
    def _repr_html_(self):
        html = ["<table style='font-family: monospace'>"]
        for row in self:
            html.append("<tr>")
            
            for col in row:
                html.append("<td>{0}</td>".format(col))
            
            html.append("</tr>")
        html.append("</table>")
        return ''.join(html)

    
class ListPara(list):
    """Overridden list class which takes a 1-dimensional list of
    strings, and renders an HTML paragraph block in the ipynb.
    Adds line breaks between each string
    Adapted from: http://calebmadrigal.com/display-list-as-table-in-ipython-notebook/
    """
    
    def _repr_html_(self):
        html = ["<p style='font-size: 14pt; font-family: times'>"]
        for line in self:
            html.append("{0}<br>".format(line))
        html.append("</p>")
        return ''.join(html)
```

# 2. Region locations, radial profile fits and extracted FWHMs in all bands, and extracted spectra on rim and downstream of rim

Add functions for calculating $m_E$, and compute $m_E$ pointwise wherever possible.  Ad hoc patch...

```python
def m_expt(e1, e2, w1, w2):
    """Calculated following Ressler et al. [2014]"""
    return np.log(w2/w1) / np.log(e2/e1)

def m_expt_err(e1, e2, w1, w2, w1e, w2e):
    """Propagate filament width errors in quadrature"""
    me = m_expt(e1, e2, w1, w2)
    return np.abs(me * 1./np.log(e2 / e1) * np.sqrt((w1e/w1)**2 + (w2e/w2)**2))

for n in regions.keys():
    reg = regions[n]
    for lab1, lab2 in zip(labels[:-1], labels[1:]):
        e1, e2 = lowerE[lab1], lowerE[lab2]
        
        fwhm1 = reg[lab1]['meas']['fwhm']
        fwhm2 = reg[lab2]['meas']['fwhm']
        fwhm1e = np.mean(reg[lab1]['meas']['fwhm-errs'])
        fwhm2e = np.mean(reg[lab2]['meas']['fwhm-errs'])

        if np.isfinite(fwhm1) and np.isfinite(fwhm2):
            reg[lab2]['meas']['m_e'] = (m_expt(e1, e2, fwhm1, fwhm2),
                                m_expt_err(e1, e2, fwhm1, fwhm2, fwhm1e, fwhm2e))
        else:
            reg[lab2]['meas']['m_e'] = float('NaN'), float('NaN')
```
