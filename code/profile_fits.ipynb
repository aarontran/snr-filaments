### Fitting functions

First, we need a profile function.  _Ressler et al._ [2014, _in press_] used a piecewise function to fit the upstream and downstream sides of the rim profiles:

\begin{align*}
    h_{\mathrm{up}}(x) &= A_u \exp \left(\frac{x_0-x}{w_u}\right) + C_u \\
    h_{\mathrm{down}}(x) &= A_d \exp \left(\frac{x - x_0}{w_d}\right)
                          + B \exp \left( \frac{-(x-x_1)^2}{2\pi \sigma^2} \right)
                          + C_d
\end{align*}

Some outstanding questions -- how do you decide the location of up/down-stream?  Is this part of the fitting procedure?
The peak intensity may be located at either $x_0$ or $x_1$ -- okay, seems physically reasonable given projection effects.  But how to decide where to split the fit.

```python
def ressler_function(x, x0, x1, Cu, Cd, wu, wd, sigma, Au, B):
    """Big fat piecewise function with a million parameters to try fitting.
    Note -- the functions should be piecewise continuous -- a constraint that has to be enforced!
    Inputs (all floats):
        x = radial position (arcsec.)
        Au, x0, wu, Cu = parameters for upstream function
        xs = location of upstream/downstream divide
        Ad, wd, B, x1, sigma, Cd = parameters for downstream function
    Output (float): value of function at radial position x
    """
    Ad = ressler_function_Ad(x0, x1, Cu, Cd, sigma, Au, B)
    def ressler_help(x, x0, x1, Cu, Cd, wu, wd, sigma, Au, B, Ad):
        if x > x0:  # Upstream, ahead of shock
            return Au * np.exp((x0-x)/wu) + Cu
        else:
            return Ad * np.exp((x-x0)/wd) + B * np.exp( -(x-x1)**2 / (2*np.pi*sigma**2)) + Cd
    ressler_vec = np.vectorize(ressler_help)
    return ressler_vec(x, x0, x1, Cu, Cd, wu, wd, sigma, Au, B, Ad)

def ressler_function_Ad(x0, x1, Cu, Cd, sigma, Au, B):
    return Au + Cu - Cd - B * np.exp( -(x0-x1)**2 / (2*np.pi*sigma**2))
```

### Two exponential function

I first try a continuous piecewise two exponential function of form:

\begin{align*}
    h_{\mathrm{up}}(x) &= A_u \exp \left(\frac{x_0-x}{w_u}\right) + C_u \\
    h_{\mathrm{down}}(x) &= A_d \exp \left(\frac{x - x_0}{w_d}\right) + C_d
\end{align*}

where the upstream/downstream split occurs at $x=x_s$ and $x_s$ is used as a fit parameter instead of $A_u$.  Once $x_s$ is determined, $A_u$ is constrained to be:

\begin{equation*}
    A_u = \exp\left(\frac{x_s - x_0}{w_u}\right) \left[ A_d \exp\left(\frac{x_s - x_0}{w_d}\right) + C_d - C_u \right]
\end{equation*}

Alternately we may constrain $A_d$ and use $A_u$ as a fit parameter:

\begin{equation*}
    A_d = \exp\left(\frac{x_0 - x_s}{w_d}\right) \left[ A_u \exp\left(\frac{x_0 - x_s}{w_u}\right) + C_u - C_d \right]
\end{equation*}

```python
def two_exp_function(x, xs, x0, Cu, Cd, wu, wd, Au):
    """Two exponential profile function, vectorized in x
    downstream amplitude Ad set by the split xs + continuity requirement
    """
    Ad = two_exp_function_Ad(xs, x0, Cu, Cd, wu, wd, Au)
    def two_exp_help(x, xs, x0, Cu, Cd, wu, wd, Au, Ad):
        if x > xs:  # Upstream, ahead of shock
            return Au * np.exp((x0-x)/wu) + Cu
        else:
            return Ad * np.exp((x-x0)/wd) + Cd
    two_exp_vec = np.vectorize(two_exp_help)
    return two_exp_vec(x, xs, x0, Cu, Cd, wu, wd, Au, Ad)

def two_exp_function_Au(xs, x0, Cu, Cd, wu, wd, Ad):
    """Value of Au set by split location xs + continuity requirement for two exponential function"""
    return np.exp((xs-x0)/wu) * ( Ad * np.exp((xs-x0)/wd) + Cd - Cu )

def two_exp_function_Ad(xs, x0, Cu, Cd, wu, wd, Au):
    """Value of Ad set by split location xs + continuity requirement for two exponential function"""
    return np.exp((x0-xs)/wd) * ( Au * np.exp((x0-xs)/wu) + Cu - Cd )
```

### Two exponential function, decoupled (one more free parameter)

I could also let one more parameter run free in the two exponential function:

\begin{align*}
    h_{\mathrm{up}}(x) &= A_u \exp \left(\frac{x_u-x}{w_u}\right) + C_u \\
    h_{\mathrm{down}}(x) &= A_d \exp \left(\frac{x - x_d}{w_d}\right) + C_d
\end{align*}

Again, $x=x_s$ is a fit parameter for the upstream/downstream split, which can be used in place of $A_u$ or $A_d$.
Because the upstream side is generally easier to fit / the data is better, it's probably better to pin $A_d$ and let $A_u$ be a free fit parameter.

\begin{equation*}
    A_u = \exp\left(\frac{x_s - x_u}{w_u}\right) \left[ A_d \exp\left(\frac{x_s - x_d}{w_d}\right) + C_d - C_u \right]
\end{equation*}

\begin{equation*}
    A_d = \exp\left(\frac{x_d - x_s}{w_d}\right) \left[ A_u \exp\left(\frac{x_u - x_s}{w_u}\right) + C_u - C_d \right]
\end{equation*}

```python
def two_exp_function_dec(x, xs, xu, xd, Cu, Cd, wu, wd, Au):
    """Two exponential profile function, with one more free parameter, vectorized in x
    downstream amplitude Ad set by the split xs + continuity requirement
    """
    Ad = two_exp_function_dec_Ad(xs, xu, xd, Cu, Cd, wu, wd, Au)
    def two_exp_help(x, xs, xu, xd, Cu, Cd, wu, wd, Au, Ad):
        if x > xs:  # Upstream, ahead of shock
            return Au * np.exp((xu-x)/wu) + Cu
        else:
            return Ad * np.exp((x-xd)/wd) + Cd
    two_exp_vec = np.vectorize(two_exp_help)
    return two_exp_vec(x, xs, xu, xd, Cu, Cd, wu, wd, Au, Ad)

def two_exp_function_dec_Au(xs, xu, xd, Cu, Cd, wu, wd, Ad):
    """Value of Au set by split location xs + continuity requirement for two exponential function"""
    return np.exp((xs-xu)/wu) * ( Ad * np.exp((xs-xd)/wd) + Cd - Cu )

def two_exp_function_dec_Ad(xs, xu, xd, Cu, Cd, wu, wd, Au):
    """Value of Ad set by split location xs + continuity requirement for two exponential function"""
    return np.exp((xd-xs)/wd) * ( Au * np.exp((xu-xs)/wu) + Cu - Cd )
```

### Two exponential function, split at $x_0$

Now instead of using a free split location $x_s$, try throwing that away and using $x_0$ as the split location (following what Satoru did?):

\begin{align*}
    h_{\mathrm{up}}(x) &= A_u \exp \left(\frac{x_0-x}{w_u}\right) + C_u \\
    h_{\mathrm{down}}(x) &= A_d \exp \left(\frac{x - x_0}{w_d}\right) + C_d
\end{align*}

where the upstream/downstream split occurs at $x=x_0$.  Continuity then allows us constrain one of $\{A_u, A_d, C_u, C_d\}$.
Consistent with above, let's constrain $A_d$ and let $A_u$ run free (try changing this up as well).

\begin{align*}
    A_d &= A_u + (C_u - C_d) \\
    A_u &= A_d + (C_d - C_u)
\end{align*}

```python
def two_exp_function_simp(x, x0, Cu, Cd, wu, wd, Au):
    """Two exponential profile function, vectorized in x
    downstream amplitude Ad set by the split xs + continuity requirement
    """
    Ad = two_exp_function_simp_Ad(Cu, Cd, Au)
    def two_exp_help(x, x0, Cu, Cd, wu, wd, Au, Ad):
        if x > x0:  # Upstream, ahead of shock
            return Au * np.exp((x0-x)/wu) + Cu
        else:
            return Ad * np.exp((x-x0)/wd) + Cd
    two_exp_vec = np.vectorize(two_exp_help)
    return two_exp_vec(x, x0, Cu, Cd, wu, wd, Au, Ad)

# Careful with the parameter signatures!
def two_exp_function_simp_Au(Cu, Cd, Ad):
    return Ad + Cd - Cu

def two_exp_function_simp_Ad(Cu, Cd, Au):
    return Au + Cu - Cd
```

### Data manipulation (smooth profiles to identify minima/maxima for initial guesses, fit domain)

A cookbook on curve fitting with SciPy: [wiki.scipy.org/Cookbook/FittingData](http://wiki.scipy.org/Cookbook/FittingData).

And, signal smoothing (maybe useful for peak/trough identification): [wiki.scipy.org/Cookbook/SignalSmooth](http://wiki.scipy.org/Cookbook/SignalSmooth).

```python
def ind_first_smoothed_min(y):
    """Smooths data and returns index of first minimum to left of global max, in smoothed data
    If no such minimum is found, return 0 (i.e., don't cut the fit domain)"""
    # List of indices of relative minima in smoothed-y
    inds_relmin = sp.signal.argrelmin(std_smooth(y))[0]  # Unpack the tuple...
    if len(inds_relmin) is 0:
        return 0
    # Index of global (non-edge) maximum in y
    ind_max = ind_smoothed_max(y)
    
    # Want to find the largest value in inds_relmin, that's smaller than ind_max
    max_left_min = 0
    for ind_min in inds_relmin:
        if ind_min > max_left_min and ind_min < ind_max:
            max_left_min = ind_min
    
    return max_left_min


def ind_smoothed_max(y):
    """Smooths data and returns index of global maximum in smoothed data"""
    ysmooth = std_smooth(y)
    # Get indices of relative maxima
    inds_relmax = sp.signal.argrelmax(ysmooth)[0]  # Unpack the tuple...
    # Get index of the largest relative maximum
    best_ind = inds_relmax[0]
    for ind in inds_relmax[1:]:
        if ysmooth[best_ind] < ysmooth[ind]:
            best_ind = ind
    return best_ind


def std_smooth(x):
    """Standardized smoothing -- SMOOTHING WINDOW CHOICE AFFECTS FIT!!!!!"""
    wlen = 21  # Arbitrary pick
    xsmooth = smooth(x, window_len=wlen, window='hanning')
    padlen = (len(xsmooth) - len(x))/2  # Should be an integer
    return xsmooth[padlen:-padlen]


def smooth(x,window_len=11,window='hanning'):
    """stolen from scipy cookbook (reimplement whatever later)"""
    if window_len < 3:
        return x
    if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
        raise ValueError, "Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'"
    
    s = numpy.r_[x[window_len-1:0:-1],x,x[-1:-window_len:-1]]
    if window == 'flat': #moving average
        w = numpy.ones(window_len,'d')
    else:
        w = eval('numpy.{}(window_len)'.format(window))
    
    y=numpy.convolve(w/w.sum(), s, mode='valid')
    return y
```

### Routines to actually perform profile fit

Fit using two exponential model (with shared parameter $x_0$).

```python
def fit_profile_2exp(x, y, y_err, lab):
    """FIT THE PROFILE"""
    
    # Set fit domain, identify smoothed peak
    leftcut = ind_first_smoothed_min(y)
    x_fit, y_fit, y_err_fit = x[leftcut:], y[leftcut:], y_err[leftcut:]
    print 'Left-of-peak smoothed minimum at: {}'.format(x[leftcut])
    x_of_max = x[ind_smoothed_max(y)]
    print 'Global smoothed maximum at: {}'.format(x_of_max)
    
    ###################
    # First fit attempt
    ###################
    
    # Freeze: xs, x0
    xs_fr = x_of_max + 1.5  # Smoothing shifts peak slightly left
    x0_fr = xs_fr + 0.5  # xs left of x0 in good fit (empirically speaking)
    
    # Free parameters: Cu, Cd, wu, wd, Au
    Cu_g = np.mean(y_fit[-3:])  # The obvious guess
    Cd_g = np.mean(y_fit[:3]) / 2.  # Usually shoots under trough
    wu_g = 0.5  # Upstream width
    wd_g = 2.5  # Downstream width
    Au_g = np.amax(y_fit)/3.  # Upstream amplitude
    init_guess = [Cu_g, Cd_g, wu_g, wd_g, Au_g]
    
    # Show initial guesses for comparison
    print ('{} initial guesses: xs = {:5.4g}, x0 = {:5.4g}, Cu = {:5.4g}, Cd = {:5.4g}, '
           'wu = {:5.4g}, wd = {:5.4g}, Au = {:5.4g}').format(lab, xs_fr, x0_fr, *init_guess)

    # Perform fit!
    fit_func = lambda xval, Cu,Cd,wu,wd,Au: two_exp_function(xval, xs_fr,x0_fr, Cu,Cd,wu,wd,Au)
    popt, pcov = sp.optimize.curve_fit(fit_func, x_fit, y_fit, p0=init_guess,
                                       sigma=y_err_fit, absolute_sigma=True)
    print ('{} fit (xs,x0 frozen): xs = {:5.4g}, x0 = {:5.4g}, Cu = {:5.4g}, Cd = {:5.4g}, '
           'wu = {:5.4g}, wd = {:5.4g}, Au = {:5.4g}').format(lab, xs_fr, x0_fr, *popt)
    
    ####################
    # Second fit attempt
    ####################
    
    # Unfreeze: xs, x0, Cu, Cd
    # All parameters free
    init_guess = np.concatenate(([xs_fr, x0_fr], popt))
    
    # Perform fit!
    fit_func = two_exp_function
    popt, pcov = sp.optimize.curve_fit(fit_func, x_fit, y_fit, p0=init_guess, sigma=y_err_fit, absolute_sigma=True)
    print ('{} fit (xs,x0,Cu,Cd thawed): xs = {:5.4g}, x0 = {:5.4g}, Cu = {:5.4g}, Cd = {:5.4g}, '
           'wu = {:5.4g}, wd = {:5.4g}, Au = {:5.4g}').format(lab, *popt)
    
    #######################
    # Print fit information
    #######################
    print ''
    np.set_printoptions(precision=2)
    print 'Errors: {}'.format(np.sqrt(np.diag(pcov)))
    
    chi2 = np.sum(((y_fit - fit_func(x_fit, *popt)) / y_err_fit)**2)
    chi2red = chi2 / (len(y_fit) - 7)  # Seven free parameters
    print 'Reduced chi-squared: {}'.format(chi2red)

    print 'Covariance matrix, variables: ' + str(['xs','x0','Cu','Cd','wu','wd','Au'])
    np.set_printoptions(precision=1)
    print pcov
    np.set_printoptions(precision=8)  # Restore to default
    
    return popt, pcov, two_exp_function, leftcut
```

```python
def fit_profile_ressler(x, y, y_err, lab):
    """Attempt to fit data to function from Ressler et al. [2014]"""
    
    # Set fit domain, identify smoothed peak
    leftcut = ind_first_smoothed_min(y)
    x_fit, y_fit, y_err_fit = x[leftcut:], y[leftcut:], y_err[leftcut:]
    print 'Left-of-peak smoothed minimum at: {}'.format(x[leftcut])
    x_of_max = x[ind_smoothed_max(y)]
    print 'Global smoothed maximum at: {}'.format(x_of_max)
    
    # CALL SIGNATURE: ressler_function(x, x0, x1, Cu, Cd, wu, wd, sigma, Au, B)
    
    ###################
    # First fit attempt
    ###################
    
    # Freeze: x0, Cu, Cd
    x0_fr = x_of_max + 1.5
    Cu_fr = np.mean(y_fit[-3:])  # The obvious guess
    Cd_fr = np.mean(y_fit[:3]) / 3.  # Usually shoots under trough
    
    # Free parameters: Cu, Cd, wu, wd, Au
    x1_g = x_of_max - 1.5  # Gaussian center
    wu_g = 0.5  # Upstream width
    wd_g = 2.5  # Downstream width
    sigma_g = 2  # Gaussian width
    Au_g = np.amax(y_fit)/1.  # Upstream amplitude
    B_g = Au_g / 6.
    init_guess = [x1_g, wu_g, wd_g, sigma_g, Au_g, B_g]

    # Perform fit!
    fit_func = lambda xval, x1,wu,wd,sigma,Au,B: ressler_function(xval ,x0_fr, x1,Cu_fr,Cd_fr,wu,wd,sigma,Au,B)
    popt, pcov = sp.optimize.curve_fit(fit_func, x_fit, y_fit, p0=init_guess,
                                       sigma=y_err_fit, absolute_sigma=True, maxfev=8000)
    allparams = np.concatenate(([x0_fr], [popt[0]], [Cu_fr, Cd_fr], popt[1:]))
    print ('{} fit (x0,Cu,Cd frozen): x0 = {:5.4g}, x1 = {:5.4g}, Cu = {:5.4g}, Cd = {:5.4g}, '
           'wu = {:5.4g}, wd = {:5.4g}, sigma = {:5.4g}, Au = {:5.4g}, B = {:5.4g}').format(lab, *allparams)
    
    ####################
    # Second fit attempt
    ####################
    
    # Unfreeze: x0
    # Leave Cu, Cd frozen
    init_guess = np.delete(allparams, [2,3])
    
    # Perform fit!
    fit_func = lambda xval, x0,x1,wu,wd,sigma,Au,B: ressler_function(xval ,x0,x1, Cu_fr,Cd_fr, wu,wd,sigma,Au,B)
    popt, pcov = sp.optimize.curve_fit(fit_func, x_fit, y_fit, p0=init_guess, sigma=y_err_fit, absolute_sigma=True, maxfev=8000)
    allparams = np.insert(popt, 2, [Cu_fr, Cd_fr])
    print ('{} fit (x0 thawed): x0 = {:5.4g}, x1 = {:5.4g}, Cu = {:5.4g}, Cd = {:5.4g}, '
           'wu = {:5.4g}, wd = {:5.4g}, sigma = {:5.4g}, Au = {:5.4g}, B = {:5.4g}').format(lab, *allparams)
    
    ###################
    # Third fit attempt
    ###################
    
    # Unfreeze: Cu
    # Leave Cd frozen
    init_guess = np.delete(allparams, 3)
    
    # Perform fit!
    fit_func = lambda xval, x0,x1,Cu,wu,wd,sigma,Au,B: ressler_function(xval ,x0,x1,Cu, Cd_fr, wu,wd,sigma,Au,B)
    popt, pcov = sp.optimize.curve_fit(fit_func, x_fit, y_fit, p0=init_guess, sigma=y_err_fit, absolute_sigma=True, maxfev=8000)
    allparams = np.insert(popt, 3, Cd_fr)
    print ('{} fit (x0 thawed): x0 = {:5.4g}, x1 = {:5.4g}, Cu = {:5.4g}, Cd = {:5.4g}, '
           'wu = {:5.4g}, wd = {:5.4g}, sigma = {:5.4g}, Au = {:5.4g}, B = {:5.4g}').format(lab, *allparams)
    
    ####################
    # Fourth fit attempt
    ####################
    # In lieu of final fit with all free parameters
    init_guess = allparams
    fit_func = ressler_function
    popt, pcov = sp.optimize.curve_fit(fit_func, x_fit, y_fit, p0=init_guess, sigma=y_err_fit, absolute_sigma=True, maxfev=8000)
    allparams = popt
    print ('{} fit (x0 thawed): x0 = {:5.4g}, x1 = {:5.4g}, Cu = {:5.4g}, Cd = {:5.4g}, '
           'wu = {:5.4g}, wd = {:5.4g}, sigma = {:5.4g}, Au = {:5.4g}, B = {:5.4g}').format(lab, *allparams)
    
    #######################
    # Print fit information
    #######################
    print ''
    np.set_printoptions(precision=2)
    print 'Errors: {}'.format(np.sqrt(np.diag(pcov)))
    
    chi2 = np.sum(((y_fit - fit_func(x_fit, *popt)) / y_err_fit)**2)
    chi2red = chi2 / (len(y_fit) - 9)  # NINE free parameters
    print 'Reduced chi-squared: {}'.format(chi2red)

    print 'Covariance matrix, variables: ' + str(['xs','x0','Cu','Cd','wu','wd','Au'])
    np.set_printoptions(precision=1)
    print pcov
    np.set_printoptions(precision=8)  # Restore to default
    
    return popt, pcov, ressler_function, leftcut  ## UPDATE THIS THINGGGLJKSDFLKJLSF
```

## Individual profile FWHM work

For each profile (every band, every region), we need to:
1. calculate FWHM
2. estimate uncertainties on FWHM
3. check that FWHM uncertainties are reasonable (i.e., $\chi^2$ space is not too weird)
4. give information to allow plotting/visualization of uncertainties.

```python
def get_fwhm(a, b, f, pars, x_max=None):
    """Manually calculate FWHM given a bracketing interval, location of maximum, and function.
    Assumptions -- there exists a single local maximum, from which the function
    MONOTONICALLY DECREASES on both sides
    
    Inputs:
        a, b (float): bounding interval
        f (function): scalar function in first argument, with form f(x, *popt)
        pars (list): list of parameters for function f
        x_max (float): location of maximum if known (skip maximization step)
    
    Output:
        3-tuple of FWHM and left, right locations of FWHM measurement
    """
    if x_max is None:
        # Multiply by -1 to maximize
        res = sp.optimize.minimize_scalar(lambda x: -1*f(x,*pars), bounds=(a,b), method='Bounded')
        x_max = res.x
    y_hm = f(x_max, *pars) / 2.
    y_lims = f(np.array([a,b]), *pars)  # Assumes that f is vectorized in x
    
    if any(y_lims >= y_hm):  # Function must fall below half max on both sides
        print 'WARNING: cannot calculate FWHM, peak not resolved in interval'
        return -1, float('NaN'), float('NaN')

    fwhm_left = sp.optimize.brentq(lambda x: f(x, *pars) - y_hm, a, x_max)
    fwhm_right = sp.optimize.brentq(lambda x: f(x, *pars) - y_hm, x_max, b)
    return fwhm_right - fwhm_left, fwhm_left, fwhm_right

def chi2red(x, y, y_err, f, pars):
    """Reduced chi-squared for a scalar function fitted to data with y-errors
    
    N.B. for a non-linear model fit, assigning 1 degree of freedom
    to each parameter may be an OVERestimate?
    (there may be other caveats/issues at hand)
    
    Inputs:
        x, y, y_err (iterable): the usual suspects
        f (function): function with call signature f(x, *pars)
        pars (list):  list of parameters for function f
    Output:
        reduced chi-squared value (float)
    """
    chi2 = np.sum( ((y - f(x, *pars))/y_err)**2 )
    return chi2 / (len(y) - len(popt))
```

Stretching functionality

```python
def stretch_x(x, xi, x0, xconst=50):
    """Parabolic stretch centered on x=x0, fixed points at x=0, x=x0
    Follows Ressler et al. [2014]
    
    Constant here is 50 arcsec instead of 200 arcsec.
    Doesn't matter, mainly scales the input range of xi
    
    No error checking -- you put bad xi, you get bad stretch
    
    Inputs:
        x (float / iterable): original x-values
        xi (float): stretch factor
        x0 (float): fixed point of stretch
    Output:
        stretched version of x (float/iterable depending on input)
    """
    return x * (1 + xi * (x - x0)/(xconst - x0))

def safe_stretch(xi, x0, xconst=50):
    """Check if xi is in the safe range
    I.e., the range in which
    1. stretch works as desired on interval [0, xconst]
    2. stretch is 1-to-1 on said interval
    """
    if x0 > xconst:
        raise Exception('Invalid stretch x0={} > xconst'.format(x0))
    return abs(xi) < (xconst - x0)/x0

def inv_stretch_x(xs, xi, x0, xconst=50):
    """Calculates the desired value of x, given a stretched x
    Returns the "sensible" value for our stretch
    
    If a=xi >0, then positive root is bigger, positive (2nd root)
    If a=xi <0, then positive root is smaller (1st root)
    
    Can return a NaN value if stretch cannot be inverted
    (physical meaning: stretch breaks down at large distances)
    """
    a = xi
    b = xconst - x0*(1 + xi)
    c = -(xconst - x0) * xs
    d = b**2 - 4*a*c
    posroot = (-b + np.sqrt(d)) / (2*a)
    
    if not safe_stretch(xi, x0, xconst):
        if xi < 0:
            raise Exception('xi={} < {}'.format(xi))
        else:
            print 'Warning: possibly bad inverse stretch (xi too large)'
    
    if xi == 0:
        return xs
    return posroot

def stretch_func(f, xi, x_max):
    """Stretched version of a function with call signature f(x, *pars)
    If function f is vectorized in x, output function will also be vectorized"""
    return lambda x, *pars: f(stretch_x(x, xi, x_max), *pars)
```

```python
from nose.tools import assert_almost_equal
from nose.tools import assert_equal

def gaussian(x, x0, sigma):
    """Not normalized, just for testing.
    mathworld.wolfram.com/GaussianFunction.html
    Expected FWHM is 2*sqrt(2 ln(2))*sigma
    endpoints at x0 +/- sqrt(2 ln(2))*sigma (~ x0 +/- 1.18*sigma)
    """
    return np.exp( -(x-x0)**2 / (2*sigma**2) )


def test_FWHM_gaussian():
    """Check FWHMs of various gaussians"""
    
    x0_vals = [-2., 0., 1., 1e4]
    sigma_vals = [0.02, 1, 1.5, 3.]
    
    for x0, sigma in zip(x0_vals, sigma_vals):
        fwhm_ans = 2*np.sqrt(2*np.log(2))*sigma
        
        # x_max not specified
        fwhm, fwhm_a, fwhm_b = get_fwhm(x0-1.2*sigma, x0+1.2*sigma, gaussian, [x0, sigma])
        
        assert_almost_equal(fwhm, fwhm_ans)
        assert_almost_equal(fwhm_a, x0 - fwhm_ans/2.)
        assert_almost_equal(fwhm_b, x0 + fwhm_ans/2.)
        
        # x_max not specified, with crappy search bounds
        fwhm, fwhm_a, fwhm_b = get_fwhm(x0-100*sigma, x0+1.1776*sigma, gaussian, [x0, sigma])
        
        assert_almost_equal(fwhm, fwhm_ans)
        assert_almost_equal(fwhm_a, x0 - fwhm_ans/2.)
        assert_almost_equal(fwhm_b, x0 + fwhm_ans/2.)
        
        # x_max specified
        fwhm, fwhm_a, fwhm_b = get_fwhm(x0-2*sigma, x0+2*sigma, gaussian, [x0, sigma], x_max=x0)
        
        assert_almost_equal(fwhm, fwhm_ans)
        assert_almost_equal(fwhm_a, x0 - fwhm_ans/2.)
        assert_almost_equal(fwhm_b, x0 + fwhm_ans/2.)
        
        # Does limit checking work?
        fwhm, fwhm_a, fwhm_b = get_fwhm(-1*sigma, 1*sigma, gaussian, [x0, sigma])
        assert_equal(fwhm, -1)


def test_stretch_function():
    """Simple checks for stretch function (not comprehensive)"""
    
    # No stretch occurs for xi = 0
    assert_equal(stretch_x(-15, 0, 10), -15)
    assert_equal(stretch_x(7, 0, 10), 7)
    
    xi_vals = [1, 2, 3]
    x0_vals = [1, 10, 100, -10]
    for xi, x0 in zip(xi_vals, x0_vals):
        assert_equal(stretch_x(0, xi, x0), 0)  # Zero invariant
        assert_equal(stretch_x(x0, xi, x0), x0)  # x0 invariant

        
def test_inv_stretch_function():
    """Simple checks for inverse stretch function (not comprehensive)"""
    xi = 1.
    x0 = 25
    xc = 55
    
    for x in np.linspace(0, 55, 13):  # Check over range [0, 55]
        assert_almost_equal(x, inv_stretch_x(stretch_x(x,xi,x0,xc), xi,x0,xc))
        assert_almost_equal(x, stretch_x(inv_stretch_x(x,xi,x0,xc), xi,x0,xc))


def test_FWHM_stretched_gaussian():
    """Check FWHMs of stretched gaussians"""
    
    # NB this is sensitive to values of xi, x0_stretch, etc.
    # MUST choose sensible values or things will blow up.
    # Note that stretching is generally asymmetric
    
    # First, get unstretched Gaussian fwhm and fwhm bounds
    x0_gauss = 20
    sigma = 5
    fwhm, a, b = get_fwhm(x0_gauss-2*sigma, x0_gauss+2*sigma, gaussian, [x0_gauss, sigma])
    print fwhm, a, b
    
    # Stretch parameters
    xi = -0.4  # Negative xi widens the Gaussian
    x0_stretch = x0_gauss  # Must always be the case
    
    # Compute new FWHM by stretching old FWHM interval
    a_str = inv_stretch_x(a, xi, x0_stretch)
    b_str = inv_stretch_x(b, xi, x0_stretch)
    fwhm_str = b_str - a_str
    print fwhm_str, a_str, b_str
    
    # Compare against stretched function
    gaussian_str = stretch_func(gaussian, xi, x0_stretch)
    
    assert_almost_equal(0.5, gaussian_str(a_str, x0_gauss, sigma))
    assert_almost_equal(0.5, gaussian_str(b_str, x0_gauss, sigma))
```

```python
def get_proj_dims(regstrs):
    """Get projection dimensions from list of regions, must all be projections
    Input: list of region strings
    Output: list of 2-tuples
        1st element is length (radial distance)
        2nd element is thickness (dist. along which projection integrates)
        
    WARNING: DIMENSIONS ARE NOT THE SAME.  Length is in arcsec., thickness is in PIXELS / PHYSICAL UNITS.
    """
    pixel2arcsec = 0.5  # Chandra spatial resolution
    
    dims = []
    for line in regstrs:
        rtype, rnums, rprops = regparse.regparse(line)
        if 'projection' not in rtype:
            print 'WARNING: region not of type projection encountered'
            print 'Type was: {}'.format(rtype)
        x1,y1,x2,y2,w = rnums
        length = np.sqrt((x2-x1)**2 + (y2-y1)**2) * pixel2arcsec
        dims.append((length, w))
    
    return dims
```

```python
# Process and store information on each region
for i in xrange(n_regions):
    
    region = {}
    region['length'] = dims[i][0]  # Units: arcsec
    region['thickness'] = np.floor(dims[i][1])  # Units: PIXELS (apply floor to be consistent with DS9)
    
    fig1, axes1 = plt.subplots(1, len(labels), figsize=(5*len(labels),5))
    fig2, axes2 = plt.subplots(1, len(labels), figsize=(5*len(labels),5))
    fig3, axes3 = plt.subplots(1, len(labels), figsize=(5*len(labels),5))
    label_axes = {lab:(ax1,ax2,ax3) for (lab,ax1,ax2,ax3) in zip(labels, axes1,axes2,axes3)}
    
    # Fit and find FWHM for each band
    for lab in labels:
        # Information for each labeled band
        region[lab] = {}  # Energy band dictionary
        ax1, ax2, ax3 = label_axes[lab]  # Subplot axes for 3 sets of subplots
        
        ########################
        # Import projection data
        ########################
        infile = '{0}_{1:02d}_band_{2}.dat'.format(inroot, i+1, lab)
        ctfile = '{0}_{1:02d}_band_{2}.dat'.format(ctroot, i+1, lab)
        data = np.loadtxt(infile)
        
        x = (data[:,0] - 1) * region['length'] / (len(data[:,0]) - 1)  # Convert to arcsec
        y = data[:,1]
        
        data_cts = np.loadtxt(ctfile)
        cts = data_cts[:,1]  # Counts averaged over region thickness
        cts_err = np.sqrt(cts * region['thickness']) / region['thickness']  # More explicit calculation
        cts_err[cts_err==0] = np.sqrt(1 * region['thickness']) / region['thickness']  # Give zero counts a non-zero error
        
        # y_err = cts_err / cts * y  # BEWARE, sometimes cts are zero... in which case you're screwed.{{}}
        y, y_err = cts, cts_err
        
        print '\nLoaded data from files {}, {}'.format(os.path.basename(infile),
                                                       os.path.basename(ctfile))
        
        ##############
        # Fit profiles
        ##############
        popt, pcov, fit_func, cut = fit_profile_2exp_simp(x, y, y_err, lab)
        x_fit, y_fit, y_err_fit = x[cut:], y[cut:], y_err[cut:]
        
        ##################
        # FWHM calculation
        ##################
        
        # For background subtraction, replace `fit_func` with
        # lambda x, *pars: fix_func(x,*pars) - y_bkg
        # and define y_bkg appropriately
        
        xmax = popt[0]  # xmax = x0 for simple two-exponential fit
        fwhm, fwhm_left, fwhm_right = get_fwhm(x_fit[0], x_fit[-1], fit_func, popt, xmax)
            
        ##########################
        # FWHM uncertainty, set-up
        ##########################
        
        # Baseline chi2red
        chi2red_min = chi2red(x_fit, y_fit, y_err_fit, fit_func, popt)
        
        # Useful functions
        def chi2red_str(xi):
            """Calculate chi2red as a function of xi, for fit function.
            Input: xi (scalar), stretch parameter, may be pos. or neg.
            Output: chi2red for stretched function
            """
            return chi2red(x_fit, y_fit, y_err_fit,
                           stretch_func(fit_func, xi, xmax), popt)
        
        # Vectorize in xi for convenience
        chi2red_str = np.vectorize(chi2red_str)
        
        def chi2red_threshold(xi):
            """Threshold for chi^2 uncertainty.  Output > 0 if \Delta\chi^2_{red} > 2.7"""
            return chi2red_str(xi) - chi2red_min - 2.7
        
        ########################
        # Xi range determination
        ########################
        
        # Increasing positive xi (decreasing FWHM) uncertainty
        # Empirically, positive xi is better behaved than negative xi
        xi_max = 0.1
        while chi2red_threshold(xi_max) < 0:
            xi_max += 0.1
        xi_bound_pos = sp.optimize.brentq(chi2red_threshold, 0, xi_max)
        if not safe_stretch(xi_bound_pos, xmax):
            print 'WARNING: xi may be too large, pathological behavior may occur for small x'
        
        # Decreasing negative xi (increasing FWHM) uncertainty
        # Empirically, this can blow up quickly (stretch becomes meaningless if too far)
        xi_min = -0.1
        while chi2red_threshold(xi_min) < 0:
            xi_min -= 0.1
        xi_bound_neg = sp.optimize.brentq(chi2red_threshold, xi_min, 0)
        if not safe_stretch(xi_bound_neg, xmax):
            raise Exception('ERROR: xi is negative and too large, pathological behavior WILL OCCUR (flipping)')
        
        ##############################
        # FWHM uncertainty calculation
        ##############################
        
        # Compute stretched FWHMs by applying INVERSE TRANSFORMATION, from x' to x.
        # E.g., get x=new_fwhm_b by solving x'(x) = old_fwhm_b
        # Thus f_stretch(new_fwhm_b) = f(old_fwhm_b) = f(x'(new_fwhm_b))
        
        fwhm_min_left = inv_stretch_x(fwhm_left, xi_bound_pos, xmax)
        fwhm_min_right = inv_stretch_x(fwhm_right, xi_bound_pos, xmax)
        fwhm_min = fwhm_min_right - fwhm_min_left
        
        fwhm_max_left = inv_stretch_x(fwhm_left, xi_bound_neg, xmax)
        fwhm_max_right = inv_stretch_x(fwhm_right, xi_bound_neg, xmax)
        fwhm_max = fwhm_max_right - fwhm_max_left
        
        region[lab]['fwhm'] = fwhm, fwhm-fwhm_min, fwhm_max-fwhm
        
        if fwhm >= 0:
            print 'FWHM {}, errors -{:5.4g}, +{:5.4g}'.format(*region[lab]['fwhm'])
        
        ############################
        # Plot the data, fits, FWHM,
        # and FWHM left/right bounds
        ############################
        ax1.errorbar(x[:cut], y[:cut], yerr=y_err[:cut], fmt='rx', alpha=0.4)  # Data NOT used in fit
        ax1.errorbar(x_fit, y_fit, yerr=y_err_fit, fmt='bo', alpha=0.7)  # Data used in fit
        ax1.plot(x, std_smooth(y), '-y', alpha = 0.4)  # Smoothed data used to help fit
        
        x_m = np.linspace(x[0], x[-1], num=200)  # x values for best fit model
        ax1.plot(x_m, fit_func(x_m, *popt), '-k', alpha=0.7)  # Best fit model
        
        # Delineate FWHM, if measurable
        if fwhm >= 0:
            ax1.axvline(fwhm_left, alpha=0.5, ls='--', c='k')
            ax1.axvline(fwhm_right, alpha=0.5, ls='--', c='k')
            
        ax1.set_xlabel('Radial dist. (arcsec.)')
        ax1.set_ylabel('Est. counts ~ intensity')
        ax1.set_title(lab)
        
        #############################
        # Plot reduced chi-squared as
        # a function of stretch (xi)
        #############################
        ax2.plot(0, [chi2red_min], 'ro')
        # Positive xi values
        xi_vals = np.linspace(0, xi_max, 100)
        ax2.plot(xi_vals, chi2red_str(xi_vals), '-k.')
        # Negative xi values
        xi_vals = np.linspace(0, xi_min, 100)
        ax2.plot(xi_vals, chi2red_str(xi_vals), '-k.')
        ax2.axvline(xi_bound_pos)
        ax2.axvline(xi_bound_neg)
        ax2.axhline(chi2red_min+2.7, c='r')
        
        ax2.set_xlabel(r'Stretch factor $\xi$')
        ax2.set_ylabel(r'Reduced $\chi^2$')
        
        ##############################
        # Plot profile function spread
        # (just to help visualization)
        ##############################
        fit_func_small = stretch_func(fit_func, xi_bound_pos, xmax)
        fit_func_large = stretch_func(fit_func, xi_bound_neg, xmax)
        
        x_m = np.linspace(x_fit[0], x_fit[-1], num=200)  # x values for best fit model, over fit domain
        ax3.plot(x_m, fit_func(x_m, *popt), '-k')  # Best fit model
        ax3.plot(x_m, fit_func_small(x_m, *popt), '--b', alpha=0.7)  # Stretching is still plottable
        ax3.plot(x_m, fit_func_large(x_m, *popt), '--r', alpha=0.7)  # even if FWHM cannot be calculated
        
        if fwhm >= 0:
            ax3.axvline(fwhm_left, alpha=0.5, ls='-', c='k')
            ax3.axvline(fwhm_right, alpha=0.5, ls='-', c='k')
            ax3.axvline(fwhm_min_left, alpha=0.5, ls='--', c='b')
            ax3.axvline(fwhm_min_right, alpha=0.5, ls='--', c='b')
            ax3.axvline(fwhm_max_left, alpha=0.5, ls='--', c='r')
            ax3.axvline(fwhm_max_right, alpha=0.5, ls='--', c='r')
        
        ax3.set_xlabel('Radial dist. (arcsec.)')
        ax3.set_ylabel('Est. counts ~ intensity')
    
    # Show the full array of subplots
    plt.show()
    
    # Plot FWHM dependence for all bands (single region)
    plt.figure(figsize=(5,5))
    for lab, n in zip(labels, xrange(len(labels))):
        fwhm, pos_err, neg_err = region[lab]['fwhm']
        if fwhm >= 0:
            plt.errorbar([n+1], [fwhm], yerr=[[pos_err],[neg_err]], fmt='bo')
    plt.xlim([0,len(labels)+2])
    plt.show()
    print ''
    
    # Do stuff to estimate FWHM dependence

# Compute average values of FWHM-energy scaling
```
